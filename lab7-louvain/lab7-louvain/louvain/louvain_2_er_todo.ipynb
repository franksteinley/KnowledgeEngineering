生成不同边概率的ER随机图

要求：编写一行代码，调用nx的erdos_renyi_graph函数，生成 ER 随机图，要求图中节点数 n = 20，连边概率 p = rate

首先学习NetworkX包的[erdos_renyi_graph函数](https://networkx.org/documentation/stable/reference/generated/networkx.generators.random_graphs.erdos_renyi_graph.html)文档

```python
"""
随机生成网络说明
erdos_renyi_graph(n, p) 会生成一个包含 n 个节点的 ER 随机图：
    • n —— 节点数量
    • p —— 对任意两节点之间连边的概率
在本示例中：n = 20，p 从 0.1 递增到 1.0（共 10 种），依次观察网络由稀到密的变化
"""

for i in range(1, 11):              # i = 1‥10，用于产生 10 个不同的连边概率
    rate = i / 10                   # 计算当前循环对应的连边概率 p (0.1, 0.2, …, 1.0)
    print(rate)                     # 在控制台打印当前概率，便于跟踪进度
    
    #【TODO】编写一行代码，调用nx的erdos_renyi_graph函数，生成 ER 随机图，要求图中节点数 n = 20，连边概率 p = rate
    G=nx.erdos_renyi_graph(20, rate)
    
    nx.draw(                        # 绘制网络图
        G,                          # 要绘制的图对象
        with_labels=True,           # 显示节点编号
        pos=nx.kamada_kawai_layout( # 使用 Kamada–Kawai 力引导布局计算节点坐标
            G
        ),
        node_size=500,              # 所有节点的像素直径
        alpha=0.8,                  # 节点与边整体透明度
        node_color="r",             # 节点颜色：红色
        edge_color="DeepPink"       # 边颜色：洋红
    )
    
    text = f'P={rate} Network'  # 构造标题文本，标明当前连边概率
    plt.title(text, fontsize=20)    # 设置标题与字号
    plt.show()                      # 显示当前图形窗口
```
